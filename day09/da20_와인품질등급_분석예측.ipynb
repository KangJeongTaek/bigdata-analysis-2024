````python
## T검정, 회귀분석을 위한 라이브러리 사용등록
from scipy import stats
from statsmodels.formula.api import ols, glm
````

````python
# 레드와인 필터링 조건
condition = df_total_wine['color'] == 'red'
````

````python
# 레드와인 품질만
red_wine_quality = df_total_wine[condition]['quality']
````

````python
# 화이트와인 품질만
white_wine_quality = df_total_wine[df_total_wine['color'] == 'white']['quality']
````

````python
stats.ttest_ind(red_wine_quality,white_wine_quality,equal_var=False)

# statistic  : T 검정 통계량 이 양수일 경우 모집단(rea_wine_quality)의 평균이 비교집단(white_wine_quality)의 평균보다 크다.
# pvalue =0.5 이상이면 이 검정은 신뢰할 수 없다
# df는 자유도
````

````python
# 회귀분석(Regression)
## 11개의 독립변수 요소를 다 더해서 그 결과로 종속변수 quality의 값이 결정된다
Rformula = 'quality ~ fixed_acidity + volatile_acidity + citric_acid + residual_sugar + chlorides + free_sulfur_dioxide + total_sulfur_dioxide + density + pH + sulphates + alcohol'
regression_result = ols(Rformula,data=df_total_wine).fit()

## 회귀분석 요약
regression_result.summary()
````
